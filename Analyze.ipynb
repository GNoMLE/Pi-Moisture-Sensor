```python
columns = ['ts', 'sensor', 'temp', 'moisture']
df = pd.read_csv("long-test.csv", header=None, names=columns)
df.index = pd.to_datetime(df.ts, unit='s')
```

```python
df.temp.plot()
```

```python
df[df.index > pd.to_datetime('2019-08-22')][df.moisture < 10000].rolling(100).mean().moisture.plot()
```
